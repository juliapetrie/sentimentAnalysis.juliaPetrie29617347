### My initial thought was to do the text in full to determine if the sentiment analysis based on polarity scores (available below) was either positive or negative.

```python
text = df.to_string()

def get_sentiment(text):
    blob = TextBlob(text)
    sentiment = blob.sentiment.polarity
    if sentiment > 0:
        return 'Positive'
    else:
        return 'Negative'

sentiment = get_sentiment(text)

print(sentiment)
```

### After receiving the positive as shown above, I wanted more information; I was not ready to blindly trust the positive conclusion. So I converted my dataset to a string, allowing me to go through the text file line by line instead of analyzing it as a whole.

```python
# Convert the DataFrame to a text string
text = df.to_string()

# Define a function to perform sentiment analysis on each line
def get_sentiment(text):
    blob = TextBlob(text)
    sentiment = blob.sentiment.polarity
    if sentiment > 0:
        return 'Positive'
    else:
        return 'Negative'

# Apply the get_sentiment function to each line 
sentiment = df.apply(lambda x: get_sentiment(x['Democrats1']), axis=1)

# Print the sentiment analysis for each line
sentiment
```

## Now to try again with Vader sentiment analysis tool

This analysis will be a lot faster than the original. I know what I'm looking for. First, let's import NLTK and start using Vader.

```python
# importing some tools
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer

sid = SentimentIntensityAnalyzer()
```

### As before let's start with democrats. 

#### This is the Vader sentiment analysis on democrat articles

```python
df8 = pd.read_table("C:\\Users\\julia\\Desktop\\democrats2.txt", sep='\t')
df8
```

##### This version outputs the sentences slightly different; view the 1's under each line of text

```python
df8['Democrats2'].value_counts()
```

#### Now, let's see the actual scores. Vader is calculating the occurrence of negative, neutral and positive in each sentence, unlike textblob that was outputting a general ruling for a sentence as a whole.

```python
df8['scores'] = df8['Democrats2'].apply(lambda review: sid.polarity_scores(review))

df8
```

#### Now to take the scores and normalize the numbers to generate the compound

```python
df8['compound']  = df8['scores'].apply(lambda score_dict: score_dict['compound'])

df8
```

#### Now, so we can easily visualize, I have created a compound score based on the compound rating. if c (compound) >=0.5 c = positive. if c < -0.5, it is negative; otherwise, c is neutral.

```python
df8['comp_score'] = df8['compound'].apply(lambda c: 'Positive' if c >= 0.5 else ('Negative' if c < -0.5 else 'Neutral'))

df8
```

```python
# Count the number of positive and negative scores
counts = df8['comp_score'].value_counts()

# Count the number of positive, neutral, and negative scores in consistent order (same as others)
counts = df8['comp_score'].value_counts()[['Positive', 'Neutral', 'Negative']]

# Define a list of colors for the bar plot
colors = ['green', 'orange', 'blue']

# Create a bar plot
plt.bar(counts.index, counts.values, color = colors)

# Add labels and title
plt.xlabel('Sentiment')
plt.ylabel('Count')

# Set the y-axis limit to 100
plt.ylim(top=100)
plt.title('Democrat Sentiment Analysis Results')

# Show the plot
plt.show()
```

#### My First thought on this graph is that the neutral here is WAY bigger than the neutral in textblob. We can see that they classify differently. Positive remains more than negative, so that continues to favour the hypothesis.
